### Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

### Importing the Dataset

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
 
df.head(10)
```

```python
df['Price'] = boston.target
df.head()
```

```python
pos = 1
fig = plt.figure(figsize=(16,24))
for i in df.columns:
    ax = fig.add_subplot(7,2,pos)
    pos = pos + 1
    sns.distplot(df[i],ax=ax)
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)
sns.heatmap(df.corr(),annot=True)
```

```python
cols=['LSTAT','INDUS','PTRATIO','RM','TAX','Price'];
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[cols])
df[cols] = scaler.transform(df[cols])
```

```python
def box_plot():
    pos = 1
    fig = plt.figure(figsize=(16,24))
    for i in cols:
        ax = fig.add_subplot(7,2,pos)
        pos = pos + 1
        sns.boxplot(df[i],ax=ax)
```

```python
box_plot()
```

```python
for i in cols:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    Lower_Whisker=Q1-1.5*IQR
    Upper_Whisker=Q3+1.5*IQR
    df=df[df[i]>Lower_Whisker]
    df=df[df[i]<Upper_Whisker]
```

```python

x=df[['LSTAT','INDUS','PTRATIO','RM','TAX']]
y=df['Price']
```

### Splitting the training and testing data

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.3,
                                                    random_state = 4)
  
print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

### Building Linear Regression Model

```python
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
  
y_pred = regressor.predict(xtest)
```

### True value vs predicted value  scatter plot

```python
plt.scatter(ytest, y_pred)
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value")
plt.show()
```

### Calculating MSE, Accuracy and residual error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ytest, y_pred)
```

```python
acc=regressor.score(xtest,ytest)
```

```python
residual_error=ytest-y_pred
re=sum(residual_error)
```

```python
print("Mean Square Error : ", mse)
print("Accuracy : ",acc )
print("Residual Error : ",re )
```
